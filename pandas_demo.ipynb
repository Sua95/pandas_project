.apply = In the pandas library, the apply() function is used to apply a function to each element of a Series (a column in a DataFrame) or to each row/column in a DataFrame. The function can be a built-in Python function or a user-defined function.

We can use lambda functions or normal functions to pass inside the .apply method

EVERY TIME WE ACCES A COLUMN LIKE THIS df["COLUMN_NAME"] RETURS A SERIES
IF WE ACCESS TO A FRAME LIKE THIS df.apply(function) WE ARE ACCESING TO JUST THE PROPERTYS OF THE COLUMNS, NOT TO ALL COLUMNS
TO APPLY SOME FUNCTION TO ALL THE COLUMNS AND ROWS WE CAN DO IT BY TYPING THE METHOD df.applymap(function) 

```python
import math

def to_lower(series):
    return series.int()

df["RANK"] = df["RANK"].apply(lambda x: int(x))
```

Cleaning Data:
df.dropna() <-- This will erase all the rows with NaN values we can modify how to erase
(axis = 'rows' or  'columns') <-- by default will chose erase by rows
(how = 'all' or 'any') <-- a conditional if 'any' of the values are missing or 'all' the values are missing
(subset =[column_1, column_2]) <-- The columns we are most interested to have a value, we pass a list even if it's just one column

To clean missing data that is not marked as NaN or None, with numpy we can use the following method
df.replace('string_to_replace', np.nan, inplace = True)

We also can use the method df.fillna("String to fill" or 0 <--a number)


```python
df.dropna(axis="rows", how="any", subset=["Sales"])
```
