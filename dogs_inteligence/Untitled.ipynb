```python
import pandas as pd
import numpy as np

df = pd.read_csv("dog_intelligence.csv")
size_df = pd.read_csv("AKC Breed Info.csv")


df.set_index('Breed', inplace = True)

df
```

```python
size_df.set_index('Breed', inplace = True)

size_df
```

```python
df.sort_index(inplace=True)
df.head(50)
```

```python
size_df.sort_index(inplace=True)
size_df
```

```python
df_all = pd.DataFrame(df.merge(size_df, left_index=True, right_index=True, how='outer'))

df_all
```

```python

df_all.replace('na', np.nan, inplace = True)
df_all.replace('not found', np.nan, inplace = True)

df_all
```

```python
df_all["obey"] = df_all["obey"].astype(float)
df_all["weight_high_lbs"] = df_all["weight_high_lbs"].astype(float)
df_all["weight_low_lbs"] = df_all["weight_low_lbs"].astype(float)
df_all["height_low_inches"] = df_all["height_low_inches"].astype(float)
df_all["height_high_inches"] = df_all["height_high_inches"].astype(float)
df_all.dtypes
```

```python

df_all["weight_mean"] = (df_all["weight_low_lbs"] + df_all["weight_high_lbs"]) / 2
df_all["height_mean"] = (df_all["height_low_inches"] + df_all["height_high_inches"]) / 2

df_all
```

```python
df_all["obey"] = df_all["obey"].str.replace("%","")
```

```python
df_all = df_all.dropna()
df_all
```

```python
archive = df_all[["weight_mean","height_mean","obey"]]
archive
archive.to_excel("dogs.xlsx", index=True)
```

```python
correl = df_all["obey"].corr(df_all["weight_mean"])
correl
```
