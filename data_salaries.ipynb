```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("ds.salaries.csv")

df
```

```python
df_mi = df.loc[(df["experience_level"] == "MI")]
df_se = df.loc[(df["experience_level"] == "SE")]
df_en = df.loc[(df["experience_level"] == "EN")]
df_ex = df.loc[(df["experience_level"] == "EX")]
```

```python
print(round(df_mi["salary_in_usd"].mean(), 2), round(df_mi["salary_in_usd"].std(),2))
print(round(df_se["salary_in_usd"].mean(), 2), round(df_se["salary_in_usd"].std(),2))
print(round(df_en["salary_in_usd"].mean(), 2), round(df_en["salary_in_usd"].std(),2))
print(round(df_ex["salary_in_usd"].mean(), 2), round(df_ex["salary_in_usd"].std(),2))
```

```python
# plot the histogram
plt.hist(df_mi['salary_in_usd'], bins=5, density=True, alpha=0.5)

# plot the KDE curve
sns.kdeplot(df_mi['salary_in_usd'], bw_method=0.20, color='red')

# add labels and title
plt.xlabel('Amount')
plt.ylabel('Salary')
plt.title('MI Salary')

# show the plot
plt.show()
```
